```python
ax=sns.countplot(x='Size' ,data=me)
```

```python
ax=sns.countplot(x='Size' ,data=me)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
me.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
S_Qty=me.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

sns.barplot(x='Size',y='Quantity', data=S_Qty)
```

```python
sns.countplot(data=me, x='Size',hue= 'Reason for Credit Entry')
```

```python
plt.figure(figsize=(10,5))

ax=sns.countplot(data=me, x='Size',hue= 'Reason for Credit Entry')

plt.show()
```

```python
#histogram 
me['Size'].hist() 
```

```python
me['Customer State'] = me['Customer State'].astype(str)
column_data = me['Customer State']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

```python
# Checking  Data  by using pie chart 
Entry_Check = me['Reason for Credit Entry'].value_counts()

#  Plot the pie chart
plt.pie(Entry_Check, labels=Entry_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
# Checking  Data  by using pie chart 
Entry_Check = me['Reason for Credit Entry'].value_counts()

#  Plot the pie chart
plt.pie(Entry_Check, labels=Entry_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#  Prepare data for pie chart
a1 = me['Size'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
# Prepare data for scatter plot
x_data = me['Reason for Credit Entry']  
y_data = me['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Reason for Credit Entry ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=me, x='Customer State')
plt.xlabel('Customer State')
plt.ylabel('Size')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = me['Customer State'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=me[me['Customer State'].isin(top_10_state.index)], x='Customer State')
plt.xlabel('Customer State')
plt.ylabel('Size')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```
